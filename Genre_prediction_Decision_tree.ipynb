```python
import pandas as pd
music= pd.read_csv('music.csv')   
music #gender 1 represents male and 0 represents female
```

```python
#we can use notnull() function to check whether there are null/missing values in data set
#notnull() returns True for non-missing/non-null values and False for missing/null values
music.notnull()
```

```python
#we can use isnull() function to check whether there are null/missing values in data set
#isnull() returns False for non-missing/non-null values and True for missing/null values
music.isnull()
```

```python
#preparing data
#Splitting data set into two set 
#input set - age, gender
#output (predicition) set - genre
X = music.drop(columns=['genre']) #removing the genre column from the music data set and storing it in x
X
```

```python
#storing the genre column in y
y=music['genre']
y
```

```python
#Decision Tree to build ml model to make predictions
#import decisiontreeclassifier from tree module of sci-kit learn library

from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()

#Now we have to train the model so that it learns patterns in data
model.fit(X, y) #fit method take input set (x) and output/prediction set (y)                 
```

```python
#Now we are using the predict function to predict the genre of 21 year old male & 22 year old female.
#Male gender is represented by 1 and female is represented by 0
model.predict([[21,1], [22,0]])

#Store the predict function in variable called prediction
prediction=model.predict([[21,1],[22,0]])
prediction
```

```python
#Model validation and accuracy
#The purpose of model validation is to check the accuracy and performance of the model basis on the past data
#So we will split the data set into training and testing data sets
#As a general rule of thumb, we allocate 70-80 percent of data from training and remaining for testing
#Once the model is trained, we can pass the test data set to get the predictions
#Then we can compare the predictions of test data set with actual values to calculate the accuracy

from sklearn.model_selection import train_test_split #import train_test_split from model_selection mod of sci-kit learn lib

train_test_split(X, y, test_size=0.2) #here we allocate 20% of music data set for training set

#This function returns a tuple so we can unpack it into 4 variables as below

X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.2) 

#X_train & X_test are input data sets for training and testing data sets
#y_train & y_test are output data sets for training and testing data sets

```

```python
#Now while training our model we want to pass only training data set in model.fit()

model.fit(X_train,y_train)
```

```python
#Similary we add our input test data sets in model.predict()
pred=model.predict(X_test)
pred
```

```python
#Now to compare the output predictions of the training set and the actual data output

# We'll import accuracy_score function from metric module of sci-kit learn lib

from sklearn.metrics import accuracy_score

#add y_test and pred in accuracy_score function to compare both of them and store the result in 'accuracy' variable

accuracy=accuracy_score(y_test,pred)
accuracy
```
